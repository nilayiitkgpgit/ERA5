## Reading the nc files 

## 1.v_850.nc

```python
import xarray as xr

# 🔹 Provide your file path
file_path = r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\v_850.nc"

# 🔹 Open dataset
ds = xr.open_dataset(file_path)

print("\n================ BASIC INFO ================\n")
print(ds)

print("\n================ DIMENSIONS ================\n")
for dim, size in ds.dims.items():
    print(f"{dim}: {size}")

print("\n================ COORDINATES ================\n")
for coord in ds.coords:
    print(f"\n🔹 {coord}")
    print(ds[coord].values[:10])  # print first 10 values

print("\n================ VARIABLES ================\n")
for var in ds.data_vars:
    print(f"\n🔹 Variable: {var}")
    print(ds[var])

print("\n================ ATTRIBUTES ================\n")
for attr in ds.attrs:
    print(f"{attr}: {ds.attrs[attr]}")

print("\n================ VARIABLE ATTRIBUTES ================\n")
for var in ds.data_vars:
    print(f"\n🔹 {var} attributes:")
    for attr, val in ds[var].attrs.items():
        print(f"   {attr}: {val}")
```
