## 1. clim q X anom v 

```python
import xarray as xr

# 🔹 File paths
clim_q_path = r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\climatologies\clim_q_850.nc"
anom_v_path = r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\anomalies\anom_v_850.nc"

# 🔹 Output
out_path = r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\decomposted terms\qbar_vprime_850.nc"

# 🔹 Load datasets
clim_q = xr.open_dataset(clim_q_path)
anom_v = xr.open_dataset(anom_v_path)

# 🔹 Extract variables
q_bar = clim_q['q']
v_prime = anom_v['v']

# 🔹 Align
q_bar, v_prime = xr.align(q_bar, v_prime)

# 🔹 Compute
qbar_vprime = q_bar * v_prime

# 🔹 Save
qbar_vprime.to_netcdf(out_path)

print("Saved:", out_path)
```

## 7. raw q X raw v

```python
import xarray as xr

# 🔹 File paths
v_path = r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\v_850.nc"
q_path = r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\q_850.nc"

# 🔹 Output path
out_path = r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\decomposted terms\qv_850.nc"

# 🔹 Load datasets
ds_v = xr.open_dataset(v_path)
ds_q = xr.open_dataset(q_path)

# 🔹 Extract variables
v = ds_v['v']
q = ds_q['q']

# 🔹 Align (VERY IMPORTANT)
v, q = xr.align(v, q)

# 🔹 Compute raw flux
qv = q * v

# 🔹 Rename variable (clean output)
qv.name = "qv"

# 🔹 Add attributes (good practice)
qv.attrs['long_name'] = "Meridional moisture flux (q * v) at 850 hPa"
qv.attrs['units'] = "kg kg-1 m s-1"

# 🔹 Save
qv.to_netcdf(out_path)

print("Saved raw moisture flux:", out_path)
```

# Starting cross varifictation

```python
import xarray as xr
import numpy as np

# 🔹 Raw flux
qv = xr.open_dataarray(r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\decomposted terms\qv_850.nc")

# 🔹 Decomposition terms
qbar_vbar   = xr.open_dataarray(r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\decomposted terms\qbar_vbar_850.nc")
qbar_vprime = xr.open_dataarray(r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\decomposted terms\qbar_vprime_850.nc")
qprime_vbar = xr.open_dataarray(r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\decomposted terms\qprime_vbar_850.nc")

# 🔹 Nonlinear term (q'v')
ds_qvprime = xr.open_dataset(r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\decomposted terms\v_q_850.nc")
qprime_vprime = ds_qvprime['vq']
```

```python
# Rename 'time' → 'valid_time'
if 'time' in qprime_vprime.dims:
    qprime_vprime = qprime_vprime.rename({'time': 'valid_time'})
```

```python
qprime_vprime = qprime_vprime.expand_dims({'pressure_level': [850]})
```

```python
qv, qbar_vbar, qbar_vprime, qprime_vbar, qprime_vprime = xr.align(
    qv, qbar_vbar, qbar_vprime, qprime_vbar, qprime_vprime
)
```

```python
qv_reconstructed = (
    qbar_vbar +
    qbar_vprime +
    qprime_vbar +
    qprime_vprime
)
```

```python
error = qv - qv_reconstructed

print("Max absolute error:", float(abs(error).max()))
print("Mean error:", float(error.mean()))
print("Std error:", float(error.std()))
```

```python
import matplotlib.pyplot as plt

# pick a safe time index
t = int(qv.valid_time.size // 2)

plt.figure(figsize=(18,5))

plt.subplot(1,3,1)
qv.isel(valid_time=t).squeeze().plot()
plt.title("Original qv")

plt.subplot(1,3,2)
qv_reconstructed.isel(valid_time=t).squeeze().plot()
plt.title("Reconstructed qv")

plt.subplot(1,3,3)
error.isel(valid_time=t).squeeze().plot()
plt.title("Error")

plt.tight_layout()
plt.show()
```
