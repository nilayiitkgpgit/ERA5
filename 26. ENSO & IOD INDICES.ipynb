# Converting the DMI (IOD index) nc file into excell file 

```python
import xarray as xr
import pandas as pd

# ==================================
# USER INPUT
# ==================================
nc_file = r"C:\Users\hp\Downloads\dmi (1).nc"
output_excel = r"D:\PHD 2025~\1.SERVER_1\IOD FILES\DMI_timeseries.xlsx"

# ==================================
# READ NETCDF
# ==================================
ds = xr.open_dataset(nc_file)

# Print dataset info (optional)
print(ds)

# ==================================
# EXTRACT VARIABLES
# ==================================
time = pd.to_datetime(ds["TIME"].values)

df = pd.DataFrame({
    "Date": time,
    "DMI": ds["DMI"].values
})

# ==================================
# FORMAT DATE
# ==================================
df["Date"] = df["Date"].dt.strftime("%d-%m-%Y")

# ==================================
# SAVE TO EXCEL
# ==================================
df.to_excel(output_excel, index=False)

print(f"Excel saved:\n{output_excel}")
```
