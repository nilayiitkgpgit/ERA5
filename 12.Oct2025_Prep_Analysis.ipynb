```python
"""
NetCDF File Inspector
---------------------
Run this script in the folder containing your .nc files, or set FOLDER_PATH below.
It prints detailed info about every .nc file found.
"""

import os
import netCDF4 as nc
import numpy as np

# ── CONFIG ──────────────────────────────────────────────────────────────────
FOLDER_PATH = r"D:\PHD 2025~\MJO_Server1"          # Change to your folder path if needed, e.g. r"C:\data\nc_files"
# ────────────────────────────────────────────────────────────────────────────

def inspect_nc_file(filepath):
    print("\n" + "=" * 70)
    print(f"FILE : {os.path.basename(filepath)}")
    print(f"PATH : {filepath}")
    print(f"SIZE : {os.path.getsize(filepath) / 1024:.2f} KB")
    print("=" * 70)

    with nc.Dataset(filepath, "r") as ds:
        # ── File format ──────────────────────────────────────────────────
        print(f"\n📁 FORMAT      : {ds.file_format}")

        # ── Global attributes ────────────────────────────────────────────
        print("\n📋 GLOBAL ATTRIBUTES:")
        if ds.ncattrs():
            for attr in ds.ncattrs():
                print(f"   {attr:30s} = {getattr(ds, attr)}")
        else:
            print("   (none)")

        # ── Dimensions ───────────────────────────────────────────────────
        print("\n📐 DIMENSIONS:")
        for name, dim in ds.dimensions.items():
            unlimited = " [UNLIMITED]" if dim.isunlimited() else ""
            print(f"   {name:20s}  size = {len(dim)}{unlimited}")

        # ── Variables ────────────────────────────────────────────────────
        print("\n📊 VARIABLES:")
        for vname, var in ds.variables.items():
            print(f"\n   ▸ {vname}")
            print(f"     dtype      : {var.dtype}")
            print(f"     dimensions : {var.dimensions}")
            print(f"     shape      : {var.shape}")

            # Variable attributes
            vatts = var.ncattrs()
            if vatts:
                print(f"     attributes :")
                for a in vatts:
                    print(f"       {a:20s} = {getattr(var, a)}")

            # Sample values (first few)
            try:
                data = var[:]
                flat = data.compressed() if isinstance(data, np.ma.MaskedArray) else data.flatten()
                if flat.size > 0:
                    sample = flat[:5].tolist()
                    print(f"     sample vals: {sample}  ...")
                    if np.issubdtype(var.dtype, np.number):
                        valid = flat[~np.isnan(flat.astype(float))] if not isinstance(data, np.ma.MaskedArray) else flat
                        if valid.size > 0:
                            print(f"     min / max  : {valid.min():.4g}  /  {valid.max():.4g}")
            except Exception as e:
                print(f"     (could not read data: {e})")


def main():
    folder = os.path.abspath(FOLDER_PATH)
    nc_files = sorted([f for f in os.listdir(folder) if f.endswith(".nc")])

    if not nc_files:
        print(f"No .nc files found in: {folder}")
        return

    print(f"\nFound {len(nc_files)} NetCDF file(s) in: {folder}")
    for fname in nc_files:
        inspect_nc_file(os.path.join(folder, fname))

    print("\n" + "=" * 70)
    print("✅ Inspection complete.")
    print("=" * 70)


if __name__ == "__main__":
    main()
```
