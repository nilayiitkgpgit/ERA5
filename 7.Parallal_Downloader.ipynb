```python
import cdsapi
import os
from datetime import datetime

# ============================================
# USER SETTINGS
# ============================================

start_year = 1980
end_year   = 2026

output_dir = r"D:\PHD 2025~\PHD DOWNLOADS\PRECEPITATION\ERA5_Daily_Prep_1980-2026"
os.makedirs(output_dir, exist_ok=True)

# --------------------------------------------
# REGION [North, West, South, East]
# --------------------------------------------
area = [50, 40, 0, 120]

# --------------------------------------------
# TIME SETTINGS
# --------------------------------------------
months_all = [f"{m:02d}" for m in range(1, 13)]
days_all   = [f"{d:02d}" for d in range(1, 32)]
times      = ['00:00']

# ============================================
# CURRENT DATE (for 2026 handling)
# ============================================

current_date  = datetime.utcnow()
current_year  = current_date.year
current_month = int(current_date.strftime("%m"))
current_day   = int(current_date.strftime("%d"))

# ============================================
# INIT CDS API
# ============================================

c = cdsapi.Client()

print("🚀 Starting ERA5 yearly downloads...\n")

# ============================================
# LOOP OVER YEARS
# ============================================

for year in range(start_year, end_year + 1):

    year_str = str(year)

    output_file = os.path.join(
        output_dir,
        f"ERA5_Daily_Rainfall_{year}.nc"
    )

    # Skip if already downloaded
    if os.path.exists(output_file):
        print(f"⏩ Skipping {year} (already exists)")
        continue

    print(f"⬇️ Downloading {year}...")

    # ----------------------------------------
    # HANDLE FULL / PARTIAL YEAR
    # ----------------------------------------
    if year < current_year:
        months = months_all
        days   = days_all

    elif year == current_year:
        print(f"⚠️ {year} is partial → downloading available data only")

        months = [f"{m:02d}" for m in range(1, current_month + 1)]
        days   = [f"{d:02d}" for d in range(1, current_day + 1)]

    else:
        print(f"❌ Skipping {year} (future year)")
        continue

    # ----------------------------------------
    # BUILD REQUEST
    # ----------------------------------------
    request = {
        'product_type': 'reanalysis',
        'variable': 'total_precipitation',
        'year': year_str,
        'month': months,
        'day': days,
        'time': times,
        'format': 'netcdf'
    }

    if area is not None:
        request['area'] = area

    # ----------------------------------------
    # DOWNLOAD
    # ----------------------------------------
    try:
        c.retrieve(
            'reanalysis-era5-single-levels',
            request,
            output_file
        )
        print(f"✅ Done: {year}\n")

    except Exception as e:
        print(f"❌ Failed for {year}: {e}\n")

print("🎉 All downloads completed!")
```

# Downloading parameters needed for Moisture transport study except Total precipitation

```python
import cdsapi
import os
import xarray as xr
from datetime import datetime

# ============================================
# USER SETTINGS
# ============================================

start_year = 1980
end_year   = 2026

output_dir = r"D:\PHD 2025~\PHD DOWNLOADS\MOISTURE_(NO PRECIPITATION)\Moisture_1980-2026_NO_tp"
os.makedirs(output_dir, exist_ok=True)

# --------------------------------------------
# REGION [North, West, South, East]
# --------------------------------------------
area = [50, 40, 0, 120]

# --------------------------------------------
# TIME SETTINGS
# daily snapshot at 00:00 only (~24 GB total)
# swap to ['00:00', '12:00'] for twice-daily (~47 GB)
# --------------------------------------------
months_all = [f"{m:02d}" for m in range(1, 13)]
days_all   = [f"{d:02d}" for d in range(1, 32)]
times      = ['00:00']

# ============================================
# VARIABLE TOGGLES
# ============================================

DOWNLOAD = {
    # --- Single-level variables ---
    'total_precipitation':          False,  # already downloaded
    'mean_sea_level_pressure':      True,   # pressure systems driving moisture
    'total_column_water_vapour':    True,   # integrated moisture content
    'skin_temperature':             False,  # not needed for moisture transport
    'top_net_thermal_radiation':    False,  # not needed for moisture transport

    # --- Pressure-level variables ---
    'pressure_levels':              True,   # master switch
}

# 850 hPa only — captures bulk of low-level moisture transport
PRESSURE_LEVELS = ['850']

# Geopotential removed — not needed for moisture transport now
PLEV_VARIABLES = [
    'u_component_of_wind',    # core moisture flux
    'v_component_of_wind',    # core moisture flux
    'specific_humidity',      # q × V = moisture flux vector
]

# ============================================
# CURRENT DATE (for partial-year handling)
# ============================================

current_date  = datetime.utcnow()
current_year  = current_date.year
current_month = int(current_date.strftime("%m"))
current_day   = int(current_date.strftime("%d"))

# ============================================
# INIT CDS API
# ============================================

c = cdsapi.Client()

# ============================================
# HELPER: build month/day lists for a year
# ============================================

def get_months_days(year):
    if year < current_year:
        return months_all, days_all
    elif year == current_year:
        print(f"  ⚠️  {year} is partial → downloading available data only")
        months = [f"{m:02d}" for m in range(1, current_month + 1)]
        days   = [f"{d:02d}" for d in range(1, current_day + 1)]
        return months, days
    else:
        return None, None   # future year

# ============================================
# MAIN LOOP
# ============================================

print("🚀 Starting ERA5 moisture transport downloads...\n")
print(f"   Variables : MSLP, TCWV, U/V wind + specific humidity @ 850 hPa")
print(f"   Time step : Daily (00:00 UTC)")
print(f"   Domain    : 0–50°N, 40–120°E")
print(f"   Period    : {start_year}–{end_year}\n")

for year in range(start_year, end_year + 1):

    year_str    = str(year)
    output_file = os.path.join(output_dir, f"ERA5_MoistureTransport_{year}.nc")

    if os.path.exists(output_file):
        print(f"⏩ Skipping {year} (already exists)")
        continue

    months, days = get_months_days(year)

    if months is None:
        print(f"❌ Skipping {year} (future year)")
        continue

    print(f"⬇️  Downloading {year}...")

    tmp_files = []

    # ==========================================
    # PART 1: SINGLE-LEVEL VARIABLES
    # ==========================================

    single_vars = [v for v, enabled in DOWNLOAD.items()
                   if enabled and v not in ('pressure_levels',)]

    if single_vars:
        tmp_sfc = os.path.join(output_dir, f"_tmp_sfc_{year}.nc")

        request_sfc = {
            'product_type': 'reanalysis',
            'variable':     single_vars,
            'year':         year_str,
            'month':        months,
            'day':          days,
            'time':         times,
            'format':       'netcdf',
        }
        if area:
            request_sfc['area'] = area

        try:
            c.retrieve('reanalysis-era5-single-levels', request_sfc, tmp_sfc)
            print(f"  ✅ Single-level done (MSLP, TCWV)")
            tmp_files.append(tmp_sfc)
        except Exception as e:
            print(f"  ❌ Single-level failed: {e}")

    # ==========================================
    # PART 2: PRESSURE-LEVEL VARIABLES
    # ==========================================

    if DOWNLOAD.get('pressure_levels') and PLEV_VARIABLES:
        tmp_plv = os.path.join(output_dir, f"_tmp_plv_{year}.nc")

        request_plv = {
            'product_type':   'reanalysis',
            'variable':       PLEV_VARIABLES,
            'pressure_level': PRESSURE_LEVELS,
            'year':           year_str,
            'month':          months,
            'day':            days,
            'time':           times,
            'format':         'netcdf',
        }
        if area:
            request_plv['area'] = area

        try:
            c.retrieve('reanalysis-era5-pressure-levels', request_plv, tmp_plv)
            print(f"  ✅ Pressure-level done (U, V, Q @ 850 hPa)")
            tmp_files.append(tmp_plv)
        except Exception as e:
            print(f"  ❌ Pressure-level failed: {e}")

    # ==========================================
    # PART 3: MERGE TEMP FILES → ONE .nc
    # ==========================================

    if len(tmp_files) == 1:
        os.rename(tmp_files[0], output_file)
        print(f"  📦 Saved: ERA5_MoistureTransport_{year}.nc\n")

    elif len(tmp_files) > 1:
        try:
            datasets = [xr.open_dataset(f) for f in tmp_files]
            merged   = xr.merge(datasets, compat='override')
            merged.to_netcdf(output_file)
            for ds in datasets:
                ds.close()
            print(f"  📦 Merged & saved: ERA5_MoistureTransport_{year}.nc\n")
        except Exception as e:
            print(f"  ❌ Merge failed for {year}: {e}\n")
        finally:
            for f in tmp_files:
                if os.path.exists(f):
                    os.remove(f)

    else:
        print(f"  ⚠️  No files downloaded for {year}, skipping.\n")

print("🎉 All downloads completed!")
print(f"\n📊 Expected total size: ~24 GB across {end_year - start_year + 1} yearly files")
print(f"💡 To load all years as one dataset:\n")
print(f"   import xarray as xr, glob")
print(f"   ds = xr.open_mfdataset(sorted(glob.glob(r'{output_dir}\\ERA5_MoistureTransport_*.nc')), combine='by_coords')")
```
