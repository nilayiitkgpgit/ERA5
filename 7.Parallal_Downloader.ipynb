```python
import cdsapi
import os
from datetime import datetime

# ============================================
# USER SETTINGS
# ============================================

start_year = 1980
end_year   = 2026

output_dir = r"D:\PHD 2025~\PHD DOWNLOADS\PRECEPITATION\ERA5_Daily_Prep_1980-2026"
os.makedirs(output_dir, exist_ok=True)

# --------------------------------------------
# REGION [North, West, South, East]
# --------------------------------------------
area = [50, 40, 0, 120]

# --------------------------------------------
# TIME SETTINGS
# --------------------------------------------
months_all = [f"{m:02d}" for m in range(1, 13)]
days_all   = [f"{d:02d}" for d in range(1, 32)]
times      = ['00:00']

# ============================================
# CURRENT DATE (for 2026 handling)
# ============================================

current_date  = datetime.utcnow()
current_year  = current_date.year
current_month = int(current_date.strftime("%m"))
current_day   = int(current_date.strftime("%d"))

# ============================================
# INIT CDS API
# ============================================

c = cdsapi.Client()

print("🚀 Starting ERA5 yearly downloads...\n")

# ============================================
# LOOP OVER YEARS
# ============================================

for year in range(start_year, end_year + 1):

    year_str = str(year)

    output_file = os.path.join(
        output_dir,
        f"ERA5_Daily_Rainfall_{year}.nc"
    )

    # Skip if already downloaded
    if os.path.exists(output_file):
        print(f"⏩ Skipping {year} (already exists)")
        continue

    print(f"⬇️ Downloading {year}...")

    # ----------------------------------------
    # HANDLE FULL / PARTIAL YEAR
    # ----------------------------------------
    if year < current_year:
        months = months_all
        days   = days_all

    elif year == current_year:
        print(f"⚠️ {year} is partial → downloading available data only")

        months = [f"{m:02d}" for m in range(1, current_month + 1)]
        days   = [f"{d:02d}" for d in range(1, current_day + 1)]

    else:
        print(f"❌ Skipping {year} (future year)")
        continue

    # ----------------------------------------
    # BUILD REQUEST
    # ----------------------------------------
    request = {
        'product_type': 'reanalysis',
        'variable': 'total_precipitation',
        'year': year_str,
        'month': months,
        'day': days,
        'time': times,
        'format': 'netcdf'
    }

    if area is not None:
        request['area'] = area

    # ----------------------------------------
    # DOWNLOAD
    # ----------------------------------------
    try:
        c.retrieve(
            'reanalysis-era5-single-levels',
            request,
            output_file
        )
        print(f"✅ Done: {year}\n")

    except Exception as e:
        print(f"❌ Failed for {year}: {e}\n")

print("🎉 All downloads completed!")
```
