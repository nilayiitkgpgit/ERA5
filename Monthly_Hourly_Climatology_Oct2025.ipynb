```python
import cdsapi
import os

# =====================================================
# SETTINGS
# =====================================================

out_dir = "/media/nilay/HDD/IIT KGP 2025/PHD/ERA5_Downloads/ERA5_1980-2025_Wind_Multiple_PL_Zipped"
os.makedirs(out_dir, exist_ok=True)

years = range(1980, 2025)
months = ["10"]

pressure_levels = ["850", "500", "300", "200"]

area = [50, 60, 0, 100]

c = cdsapi.Client()

# =====================================================
# DOWNLOAD LOOP
# =====================================================

for level in pressure_levels:

    for year in years:

        out_file = os.path.join(
            out_dir,
            f"ERA5_{level}hPa_{year}.nc"
        )

        if os.path.exists(out_file):
            print("‚úÖ Skip:", level, year)
            continue

        print(f"\n‚¨á Downloading {level} hPa ‚Äî {year}")

        request = {
            "product_type": "reanalysis",
            "variable": [
                "geopotential",
                "u_component_of_wind",
                "v_component_of_wind"
            ],
            "pressure_level": level,
            "year": str(year),
            "month": months,
            "day": [f"{d:02d}" for d in range(1, 32)],
            "daily_statistic": "daily_mean",
            "time_zone": "utc+00:00",
            "frequency": "1_hourly",
            "data_format": "netcdf",
            "area": area
        }

        c.retrieve(
            "derived-era5-pressure-levels-daily-statistics",
            request,
            out_file
        )

print("\n‚úÖ ALL JOBS SUBMITTED")
```
