# Downloading daily precipitation data from 1980-2025 for 12 months => (ERA5 Land Data)

```python
import cdsapi
import os

# ============================================
# USER SETTINGS
# ============================================

start_year = 1980
end_year   = 2025

output_dir = r"D:\PHD 2025~\PHD\ERA5_Downloads\4.ERA5_1980-2025_12Months_DailyPrep"
os.makedirs(output_dir, exist_ok=True)

# --------------------------------------------
# REGION [North, West, South, East]
# --------------------------------------------
area = [50, 40, 0, 120]

# --------------------------------------------
# VARIABLE SWITCHES (True / False)
# --------------------------------------------
download_rainfall = True      # ERA5-Land total precipitation
download_winds    = False      # ERA5 pressure level winds

# --------------------------------------------
# PRESSURE LEVEL SETTINGS
# --------------------------------------------
pressure_levels = ['850']     # add more: ['850','700','500']

# --------------------------------------------
# TIME SETTINGS
# --------------------------------------------
months = [f"{m:02d}" for m in range(1, 13)]
days   = [f"{d:02d}" for d in range(1, 32)]

# For winds → use full day
times_wind = [f"{h:02d}:00" for h in range(24)]

# For rainfall → only 00:00 (daily accumulation trick)
times_rain = ['00:00']

# ============================================
# INIT CDS API
# ============================================

c = cdsapi.Client()

# ============================================
# LOOP THROUGH YEARS
# ============================================

for year in range(start_year, end_year + 1):

    print(f"\nProcessing year: {year}")

    # ========================================
    # 1. ERA5-LAND RAINFALL
    # ========================================
    if download_rainfall:

        print("  → Downloading rainfall...")

        rain_file = os.path.join(
            output_dir,
            f"ERA5_Land_Rainfall_{year}.nc"
        )

        request_rain = {
            'variable': 'total_precipitation',
            'year': str(year),
            'month': months,
            'day': days,
            'time': times_rain,
            'format': 'netcdf'
        }

        if area is not None:
            request_rain['area'] = area

        c.retrieve(
            'reanalysis-era5-land',
            request_rain,
            rain_file
        )

        print(f"  ✓ Saved rainfall: {rain_file}")

    # ========================================
    # 2. ERA5 PRESSURE LEVEL WINDS
    # ========================================
    if download_winds:

        print("  → Downloading winds...")

        wind_file = os.path.join(
            output_dir,
            f"ERA5_Winds_{year}.nc"
        )

        request_wind = {
            'product_type': 'reanalysis',
            'variable': [
                'u_component_of_wind',
                'v_component_of_wind'
            ],
            'pressure_level': pressure_levels,
            'year': str(year),
            'month': months,
            'day': days,
            'time': times_wind,
            'format': 'netcdf'
        }

        if area is not None:
            request_wind['area'] = area

        c.retrieve(
            'reanalysis-era5-pressure-levels',
            request_wind,
            wind_file
        )

        print(f"  ✓ Saved winds: {wind_file}")

print("\nAll downloads completed!")
```

```python
import cdsapi
import os
from datetime import datetime

# ============================================
# USER SETTINGS
# ============================================

start_year = 1980
end_year   = 2026   # include current year

output_dir = r"D:\PHD 2025~\PHD DOWNLOADS\PRECEPITATION"
os.makedirs(output_dir, exist_ok=True)

output_file = os.path.join(output_dir, "ERA5_365D_Daily_Rainfall_1980_2026.nc")

# --------------------------------------------
# REGION [North, West, South, East]
# --------------------------------------------
area = [50, 40, 0, 120]

# --------------------------------------------
# TIME SETTINGS
# --------------------------------------------
months = [f"{m:02d}" for m in range(1, 13)]
days   = [f"{d:02d}" for d in range(1, 32)]
times  = ['00:00']   # ~daily value

# ============================================
# HANDLE CURRENT YEAR LIMIT
# ============================================

current_date = datetime.utcnow()
current_year = current_date.year
current_month = int(current_date.strftime("%m"))
current_day   = int(current_date.strftime("%d"))

# ============================================
# BUILD REQUEST
# ============================================

years = [str(y) for y in range(start_year, end_year + 1)]

request = {
    'product_type': 'reanalysis',
    'variable': 'total_precipitation',
    'year': [],
    'month': [],
    'day': [],
    'time': times,
    'format': 'netcdf'
}

# Loop to properly handle 2026
for y in range(start_year, end_year + 1):

    y_str = str(y)

    if y < current_year:
        # full year
        request['year'].append(y_str)
        request['month'] = months
        request['day']   = days

    elif y == current_year:
        print(f"⚠️ {current_year} is partial year → downloading available data only")

        request['year'].append(y_str)
        request['month'] = [f"{m:02d}" for m in range(1, current_month + 1)]
        request['day']   = [f"{d:02d}" for d in range(1, current_day + 1)]

# Add area
if area is not None:
    request['area'] = area

# ============================================
# DOWNLOAD
# ============================================

c = cdsapi.Client()

print("🚀 Downloading ERA5 rainfall (single file)...")

c.retrieve(
    'reanalysis-era5-single-levels',
    request,
    output_file
)

print(f"✅ Download completed: {output_file}")
```
