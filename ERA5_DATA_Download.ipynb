```python
import xarray as xr

base = r"D:\PHD IIT KGP\Codes @ VSCODE\ERA5 DATA Download\extracted"

ds_inst = xr.open_dataset(f"{base}\data_stream-oper_stepType-instant.nc")
ds_acc  = xr.open_dataset(f"{base}\data_stream-oper_stepType-accum.nc")

print("Instant variables:", list(ds_inst.data_vars))
print("Accumulated variables:", list(ds_acc.data_vars))
```

```python
ds = xr.merge([ds_inst, ds_acc])
print(ds)
```

Convert precipitation to mm

```python
ds["tp_mm"] = ds.tp * 1000
```

Daily precipitation

```python
tp_daily = ds.tp_mm.resample(valid_time="1D").sum()
```

Daily mean temperature

```python
t2m_daily = ds.t2m.resample(valid_time="1D").mean()
```

Spatial averaging over WD box (classic metric)

```python
tp_box = tp_daily.mean(dim=["latitude", "longitude"])
t2m_box = t2m_daily.mean(dim=["latitude", "longitude"])
```

Convert K → °C 

```python
# Convert K → °C (ERA5 is always Kelvin)
t2m_daily_c = t2m_daily - 273.15
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

fig = plt.figure(figsize=(8, 6))
ax = plt.axes(projection=ccrs.PlateCarree())

t2m_daily_c.isel(valid_time=0).plot(
    ax=ax,
    transform=ccrs.PlateCarree(),
    cmap="coolwarm",
    cbar_kwargs={"label": "Temperature (°C)"}
)

ax.coastlines(resolution="50m")
ax.add_feature(cfeature.BORDERS, linewidth=0.5)
ax.gridlines(draw_labels=True, linewidth=0.3, linestyle="--")

ax.set_title("Daily Mean 2 m Temperature (°C)\nERA5 – 2020-01-01")

plt.tight_layout()
plt.show()
```
