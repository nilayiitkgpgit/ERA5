# Validating the raw , clim , anomally file over Kharagpur tile =>

```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ==============================
# 📂 FILE PATHS
# ==============================
raw_file  = r"/mnt/hdd4/A_Nilay/RAW_Daily_Prep_1980-2026/ERA5_merged_1980_2026_new.nc"
clim_file = r"/mnt/hdd4/A_Nilay/RAW_Daily_Prep_1980-2026/clim.nc"
anom_file = r"/mnt/hdd4/A_Nilay/RAW_Daily_Prep_1980-2026/anomaly_new.nc"

# ==============================
# 📍 TARGET LOCATION (Kharagpur)
# ==============================
TARGET_LAT = 22.3
TARGET_LON = 87.3

# ==============================
# LOAD DATA
# ==============================
ds_raw  = xr.open_dataset(raw_file)
ds_clim = xr.open_dataset(clim_file)
ds_anom = xr.open_dataset(anom_file)

raw  = ds_raw['tp']
clim = ds_clim['tp']
anom = ds_anom['tp']

# rename time if needed
if 'valid_time' in raw.dims:
    raw = raw.rename({'valid_time': 'time'})
if 'valid_time' in clim.dims:
    clim = clim.rename({'valid_time': 'time'})
if 'valid_time' in anom.dims:
    anom = anom.rename({'valid_time': 'time'})

# ==============================
# 📍 NEAREST GRID POINT
# ==============================
def nearest_idx(arr, val):
    return int(np.argmin(np.abs(arr - val)))

lat_idx = nearest_idx(raw.latitude.values, TARGET_LAT)
lon_idx = nearest_idx(raw.longitude.values, TARGET_LON)

# ==============================
# 📊 EXTRACT TIME SERIES
# ==============================
raw_ts  = raw[:, lat_idx, lon_idx]
anom_ts = anom[:, lat_idx, lon_idx]

# ==============================
# MATCH CLIMATOLOGY CORRECTLY
# ==============================
doy = raw_ts['time'].dt.dayofyear

clim_ts = clim.sel(latitude=raw.latitude[lat_idx],
                   longitude=raw.longitude[lon_idx])

# IMPORTANT FIX
clim_vals = clim_ts.values

# handle leap year safely
clim_matched = xr.DataArray(
    [clim_vals[d-1] if d <= len(clim_vals) else clim_vals[-1] for d in doy.values],
    coords={'time': raw_ts['time']},
    dims=['time']
)

# ==============================
# 🔍 COMPUTE RAW - CLIM
# ==============================
computed_anom = raw_ts - clim_matched

# ==============================
# 📊 BUILD DATAFRAME
# ==============================
df = pd.DataFrame({
    "raw": raw_ts.values,
    "clim": clim_matched.values,
    "anom_file": anom_ts.values,
    "raw_minus_clim": computed_anom.values
}, index=pd.to_datetime(raw_ts['time'].values))

# Select a shorter period for visualization
df_plot = df["2025-10-01":"2025-10-31"]

# ==============================
# 🔍 NUMERICAL CHECK
# ==============================
diff = df["anom_file"] - df["raw_minus_clim"]

print("\n===== VALIDATION =====")
print("Mean difference:", diff.mean())
print("Max difference :", diff.max())
print("Min difference :", diff.min())

# ==============================
# 📈 PLOT
# ==============================
plt.figure(figsize=(14,6))

plt.plot(df_plot.index, df_plot["raw"], label="Raw", linewidth=2)
plt.plot(df_plot.index, df_plot["clim"], '--', label="Climatology")
plt.plot(df_plot.index, df_plot["anom_file"], label="Anomaly (file)", linewidth=2)
plt.plot(df_plot.index, df_plot["raw_minus_clim"], ':', label="Raw - Clim (computed)", linewidth=2)

plt.axhline(0, color='black', linewidth=0.5)

plt.title("Validation at Kharagpur (22.3N, 87.3E)\nCheck: anomaly = raw - climatology")
plt.ylabel("Precipitation (meters)")
plt.legend()
plt.grid(alpha=0.3)

plt.show()
```

# Creating MJO file using LANCZOS filter

### 🔹 STEP 1: Setup

```python
import xarray as xr
import numpy as np
from scipy.signal import filtfilt
```

### 🔹 STEP 2: Load anomaly

```python
file = r"/mnt/hdd4/A_Nilay/RAW_Daily_Prep_1980-2026/anomaly_new.nc"

ds = xr.open_dataset(file, chunks={'time': 100})
anom = ds['tp']

# rename if needed
if 'valid_time' in anom.dims: 
    anom = anom.rename({'valid_time': 'time'})
```

### 🔹 STEP 3: Create Lanczos weights

```python
def lanczos_weights(window, cutoff):
    n = (window - 1) // 2
    w = np.zeros(window)

    for i in range(window):
        k = i - n
        if k == 0:
            w[i] = 2 * cutoff
        else:
            w[i] = np.sin(2*np.pi*cutoff*k)/(np.pi*k)
            w[i] *= np.sin(np.pi*k/n)/(np.pi*k/n)
    return w

# parameters
window = 121   # ~4 months
low_cut = 1/100
high_cut = 1/20

w_low = lanczos_weights(window, low_cut)
w_high = lanczos_weights(window, high_cut)

weights = w_high - w_low
```

### 🔹 STEP 4: Apply filter (IMPORTANT)

```python
anom = anom.chunk({'time': -1})

def apply_filter(ts, weights):
    return filtfilt(weights, [1.0], ts, axis=0, padtype='odd')

mjo = xr.apply_ufunc(
    apply_filter,
    anom,
    input_core_dims=[['time']],
    output_core_dims=[['time']],
    vectorize=True,
    dask="parallelized",
    kwargs={'weights': weights},
    output_dtypes=[float],
    dask_gufunc_kwargs={'allow_rechunk': True}
)
```

```python
print("Anomaly max:", float(anom.max().compute()))
print("MJO max:", float(mjo.max().compute()))
```

### 🔹 STEP 5: Creating only MJO file

```python
import xarray as xr

# ==============================
# ENSURE CORRECT DIMENSION ORDER
# ==============================
mjo = mjo.transpose('time', 'latitude', 'longitude')

# ==============================
# SET VARIABLE NAME + ATTRIBUTES
# ==============================
mjo.name = "tp_mjo"
mjo.attrs['long_name'] = "MJO filtered precipitation (20-100 day)"
mjo.attrs['units'] = "m"

# ==============================
# CONVERT TO DATASET (IMPORTANT)
# ==============================
ds_out = mjo.to_dataset()

# ==============================
# SAVE (OPTIMIZED)
# ==============================
encoding = {
    "tp_mjo": {
        "zlib": True,
        "complevel": 4,
        "dtype": "float32"
    }
}

ds_out.to_netcdf("lanczos_mjo_only.nc", encoding=encoding)

print("✅ MJO file created: lanczos_mjo_only.nc")
```

```python
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ==============================
# 📂 FILE PATHS
# ==============================
FOLDER = "/mnt/hdd4/A_Nilay/RAW_Daily_Prep_1980-2026/"

f_raw  = FOLDER + "ERA5_merged_1980_2026_new.nc"
f_clim = FOLDER + "clim.nc"
f_anom = FOLDER + "anomaly_new.nc"
f_mjo  = FOLDER + "lanczos_mjo_only.nc"
f_non  = FOLDER + "non_mjo.nc"

# ==============================
# 📍 KHARAGPUR LOCATION
# ==============================
TARGET_LAT = 22.3
TARGET_LON = 87.3

START = "1980-10-01"
END   = "2025-10-31"

# ==============================
# 📖 LOAD DATA
# ==============================
ds_raw  = xr.open_dataset(f_raw)
ds_clim = xr.open_dataset(f_clim)
ds_anom = xr.open_dataset(f_anom)
ds_mjo  = xr.open_dataset(f_mjo)
ds_non  = xr.open_dataset(f_non)

raw  = ds_raw['tp']
clim = ds_clim['tp']
anom = ds_anom['tp']
mjo  = ds_mjo['tp_mjo']
non  = ds_non['tp_non_mjo']

# ==============================
# 🧠 FIX DIMENSIONS
# ==============================
if 'valid_time' in raw.dims:
    raw = raw.rename({'valid_time': 'time'})
if 'valid_time' in anom.dims:
    anom = anom.rename({'valid_time': 'time'})

# ==============================
# 📍 SELECT GRID
# ==============================
raw_pt  = raw.sel(latitude=TARGET_LAT, longitude=TARGET_LON, method='nearest')
anom_pt = anom.sel(latitude=TARGET_LAT, longitude=TARGET_LON, method='nearest')
mjo_pt  = mjo.sel(latitude=TARGET_LAT, longitude=TARGET_LON, method='nearest')
non_pt  = non.sel(latitude=TARGET_LAT, longitude=TARGET_LON, method='nearest')

# ==============================
# 🕒 TIME FILTER
# ==============================
raw_pt  = raw_pt.sel(time=slice(START, END))
anom_pt = anom_pt.sel(time=slice(START, END))
mjo_pt  = mjo_pt.sel(time=slice(START, END))
non_pt  = non_pt.sel(time=slice(START, END))

# ==============================
# 📊 CLIMATOLOGY MATCHING (FIXED)
# ==============================
clim_pt = clim.sel(latitude=TARGET_LAT, longitude=TARGET_LON, method='nearest')

# Extract day-of-year from raw
doy = raw_pt['time'].dt.dayofyear.values

# Climatology has 366 days (1–366)
# Convert to index (0–365)
clim_matched = xr.DataArray(
    clim_pt.values[doy - 1],
    coords={'time': raw_pt['time']},
    dims=['time']
)

# ==============================
# 🔍 VALIDATION CHECKS
# ==============================
raw_minus_clim = raw_pt - clim_matched
anom_minus_mjo = anom_pt - mjo_pt

print("\n===== VALIDATION =====")
print("Raw - Clim vs Anomaly → Mean diff:",
      float((raw_minus_clim - anom_pt).mean()))

print("Anomaly - MJO vs Non-MJO → Mean diff:",
      float((anom_minus_mjo - non_pt).mean()))

# ==============================
# 📈 PLOT 1: RAW vs CLIM + ANOM
# ==============================
plt.figure(figsize=(14,5))

plt.plot(raw_pt['time'], raw_pt*1000, label='Raw', linewidth=2)
plt.plot(raw_pt['time'], clim_matched*1000, '--', label='Climatology')
plt.plot(raw_pt['time'], anom_pt*1000, label='Anomaly')

plt.title("Raw = Climatology + Anomaly (Kharagpur, Oct 2025)")
plt.ylabel("mm/day")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# ==============================
# 📈 PLOT 2: ANOM vs MJO + NON-MJO
# ==============================
plt.figure(figsize=(14,5))

plt.plot(anom_pt['time'], anom_pt*1000, label='Anomaly', linewidth=2)
plt.plot(mjo_pt['time'], mjo_pt*1000, label='MJO')
plt.plot(non_pt['time'], non_pt*1000, label='Non-MJO')

plt.title("Anomaly = MJO + Non-MJO (Kharagpur, Oct 2025)")
plt.ylabel("mm/day")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# ==============================
# 📈 PLOT 3: ERROR CHECK
# ==============================
error = (anom_pt - (mjo_pt + non_pt)) * 1000

plt.figure(figsize=(14,4))
plt.plot(error['time'], error, color='red')
plt.axhline(0, color='black')

plt.title("Residual Error (Should be ~0)")
plt.ylabel("mm/day")
plt.grid(alpha=0.3)
plt.show()
```

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

# ==============================
# 📂 FILE PATH
# ==============================
FILE = "/mnt/hdd4/A_Nilay/RAW_Daily_Prep_1980-2026/non_mjo.nc"

# ==============================
# 📍 USER INPUT
# ==============================
TARGET_LAT = 22.3
TARGET_LON = 87.3

START = "1980-10-01"
END   = "2025-10-31"

# ==============================
# 📖 LOAD DATA
# ==============================
ds = xr.open_dataset(FILE)
var = list(ds.data_vars)[0]   # auto-detect variable
data = ds[var]

print("Variable name:", var)
print("Units:", data.attrs.get("units", "Not defined"))

# ==============================
# 🧠 FIX DIMENSION (if needed)
# ==============================
if 'valid_time' in data.dims:
    data = data.rename({'valid_time': 'time'})

# ==============================
# 📍 SELECT GRID
# ==============================
pt = data.sel(latitude=TARGET_LAT,
              longitude=TARGET_LON,
              method='nearest')

# ==============================
# 🕒 TIME FILTER
# ==============================
pt = pt.sel(time=slice(START, END))

# ==============================
# 📊 CONVERT TO mm/day (ONLY FOR PLOT)
# ==============================
pt_mm = pt * 1000

# ==============================
# 📊 STATS
# ==============================
max_val = float(pt_mm.max())
min_val = float(pt_mm.min())

print("\n===== NON-MJO STATS =====")
print(f"Max: {max_val:.3f} mm/day")
print(f"Min: {min_val:.3f} mm/day")

# ==============================
# 📈 PLOT
# ==============================
plt.figure(figsize=(14,5))

plt.plot(pt_mm['time'], pt_mm, color='green', linewidth=2)

plt.axhline(0, color='black', linewidth=0.5)

plt.title(f"Non-MJO Rainfall at ({TARGET_LAT}, {TARGET_LON})")
plt.ylabel("mm/day")
plt.grid(alpha=0.3)

plt.show()

# ==============================
# CLOSE
# ==============================
ds.close()
```
