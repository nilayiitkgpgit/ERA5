```python
import os
import time
import threading
import calendar
import zipfile
import glob
import cdsapi

# =====================================================
# USER CONTROLS
# =====================================================

out_dir = r"D:\PHD IIT KGP\Codes @ VSCODE\ERA5 Monthly Data"

start_year = 2020
end_year   = 2024

months = list(range(1, 13))   # Jan‚ÄìDec

#area = [45, 40, 20, 85]   # India
area = None             # Global

var_short = "TP_SF_T2M"

# =====================================================
# HELPER FUNCTIONS
# =====================================================

def month_name(m):
    return calendar.month_name[m][:12]   # Jan, Feb, Mar...Dec


def monitor_download(file_path, interval=5):
    print("üì• Download started...")
    last_size = -1
    stable_count = 0

    while True:
        if os.path.exists(file_path):
            size = os.path.getsize(file_path) / (1024**2)
            print(f"   Downloaded: {size:.2f} MB", end="\r")

            if size == last_size:
                stable_count += 1
            else:
                stable_count = 0

            last_size = size

            if stable_count >= 3:
                break

        time.sleep(interval)

    print(f"\n‚úÖ Download finished: {size:.2f} MB")


# =====================================================
# MAIN DOWNLOAD LOGIC
# =====================================================

os.makedirs(out_dir, exist_ok=True)
c = cdsapi.Client()

years = [str(y) for y in range(start_year, end_year + 1)]

for m in months:

    mon_name = month_name(m)
    out_file = f"{var_short}_{mon_name}_{start_year}{end_year}.nc"
    out_path = os.path.join(out_dir, out_file)

    print(f"\nüöÄ Downloading: {out_file}")

    request = {
        "product_type": "monthly_averaged_reanalysis",
        "variable": [
            "total_precipitation",
            "snowfall",
            "2m_temperature"
        ],
        "year": years,
        "month": f"{m:02d}",
        "time": "00:00",
        "data_format": "netcdf"
    }

    if area is not None:
        request["area"] = area

    monitor_thread = threading.Thread(
        target=monitor_download,
        args=(out_path,)
    )
    monitor_thread.start()

    c.retrieve(
        "reanalysis-era5-single-levels-monthly-means",
        request,
        out_path
    )

    monitor_thread.join()

    # =================================================
    # ZIP HANDLING + SAFE RENAMING
    # =================================================

    if zipfile.is_zipfile(out_path):
        print("‚ö† ZIP detected ‚Äî extracting and renaming")

        with zipfile.ZipFile(out_path, "r") as z:
            z.extractall(out_dir)

        os.remove(out_path)

        extracted_files = glob.glob(
            os.path.join(out_dir, "data_stream-*.nc")
        )

        for f in extracted_files:
            step_type = os.path.basename(f).replace(".nc", "")
            new_name = (
                f"{var_short}_{mon_name}_{start_year}{end_year}_"
                f"{step_type}.nc"
            )
            new_path = os.path.join(out_dir, new_name)

            os.rename(f, new_path)
            print(f"‚úÖ Saved ‚Üí {new_name}")

    else:
        print("‚úÖ NetCDF delivered directly (no ZIP)")

    print("--------------------------------------------------")
```
