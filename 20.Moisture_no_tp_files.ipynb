### Plotting the time msl series=>

```python
import xarray as xr
import matplotlib.pyplot as plt

# ---- FILE PATHS ----
clim_file = r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\climatologies\clim_msl.nc"
anom_file = r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\anomalies\anom_msl.nc"

# ---- LOAD DATA ----
ds_clim = xr.open_dataset(clim_file)
ds_anom = xr.open_dataset(anom_file)

# ---- LOCATION ----
lat = 22.57
lon = 88.36

# ---- EXTRACT POINT ----
msl_clim = ds_clim['msl'].sel(latitude=lat, longitude=lon, method="nearest")
msl_anom = ds_anom['msl'].sel(latitude=lat, longitude=lon, method="nearest")

# ---- CONVERT Pa → hPa ----
msl_clim = msl_clim / 100.0
msl_anom = msl_anom / 100.0

# ---- DAY OF YEAR FOR CLIMATOLOGY ----
doy = msl_clim['valid_time'].dt.dayofyear

# ---- CREATE FIGURE ----
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# =========================
# 📈 CLIMATOLOGY PLOT
# =========================
axes[0].plot(doy, msl_clim, color='blue', linewidth=2)
axes[0].set_title("Climatological MSL (Kolkata)")
axes[0].set_xlabel("Day of Year")
axes[0].set_ylabel("MSL (hPa)")
axes[0].grid(True)

# Month ticks (optional but useful)
axes[0].set_xticks(
    [1, 32, 60, 91, 121, 152, 182, 213, 244, 274, 305, 335]
)
axes[0].set_xticklabels(
    ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
)

# =========================
# 📉 ANOMALY PLOT
# =========================
axes[1].plot(msl_anom['valid_time'], msl_anom, color='red', linewidth=1)

# zero reference line
axes[1].axhline(0, color='black', linestyle='--', linewidth=1)

axes[1].set_title("MSL Anomaly (Kolkata)")
axes[1].set_xlabel("Time")
axes[1].set_ylabel("Anomaly (hPa)")
axes[1].grid(True)

# ---- FINAL LAYOUT ----
plt.tight_layout()
plt.show()
```

# Ploting wind anomaly over Indian region

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# ---- FILE PATHS ----
u_file = r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\anomalies\anom_u_850.nc"
v_file = r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\anomalies\anom_v_850.nc"

# ---- LOAD DATA ----
ds_u = xr.open_dataset(u_file)
ds_v = xr.open_dataset(v_file)

u = ds_u['u']
v = ds_v['v']

# ---- SELECT PRESSURE LEVEL (if exists) ----
if 'pressure_level' in u.dims:
    u = u.sel(pressure_level=850)
    v = v.sel(pressure_level=850)

# ---- USER INPUT DATE RANGE ----
start_date = "2025-10-25"
end_date   = "2025-10-31"

u_sel = u.sel(valid_time=slice(start_date, end_date))
v_sel = v.sel(valid_time=slice(start_date, end_date))

# ---- GET NUMBER OF TIME STEPS ----
times = u_sel['valid_time'].values
n = len(times)

# ---- SUBPLOT GRID (AUTO) ----
ncols = min(3, n)  # max 3 columns
nrows = int(np.ceil(n / ncols))

fig, axes = plt.subplots(
    nrows, ncols,
    figsize=(5*ncols, 4*nrows),
    subplot_kw={'projection': ccrs.PlateCarree()}
)

axes = np.array(axes).flatten()

# ---- LOOP OVER TIME ----
for i, t in enumerate(times):

    ax = axes[i]

    u_t = u_sel.sel(valid_time=t)
    v_t = v_sel.sel(valid_time=t)

    # ---- SUBSET INDIA REGION ----
    u_t = u_t.sel(latitude=slice(40, 5), longitude=slice(65, 100))
    v_t = v_t.sel(latitude=slice(40, 5), longitude=slice(65, 100))

    # ---- THIN GRID (IMPORTANT FOR ARROWS) ----
    skip = 5
    u_plot = u_t[::skip, ::skip]
    v_plot = v_t[::skip, ::skip]

    lon = u_plot.longitude
    lat = u_plot.latitude

    # ---- COMPUTE WIND SPEED (for reference) ----
    speed = np.sqrt(u_plot**2 + v_plot**2)

    # ---- PLOT QUIVER ----
    q = ax.quiver(
        lon, lat,
        u_plot, v_plot,
        scale=200,        # adjust if arrows too small/large
        width=0.002,
        headwidth=3
    )

    # ---- MAP FEATURES ----
    ax.coastlines()
    ax.add_feature(cfeature.BORDERS, linestyle=':')

    ax.set_extent([65, 100, 5, 40])

    ax.set_title(str(np.datetime_as_string(t, unit='D')))

# ---- REMOVE EMPTY SUBPLOTS ----
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

# ---- ADD QUIVER KEY ----
plt.quiverkey(q, 0.9, -0.1, 5, "5 m/s", labelpos='E')

plt.tight_layout()
plt.show()
```

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# =========================
# ---- FILE PATHS ----
# =========================
u_file = r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\anomalies\anom_u_850.nc"
v_file = r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\anomalies\anom_v_850.nc"

# =========================
# ---- LOAD DATA ----
# =========================
ds_u = xr.open_dataset(u_file)
ds_v = xr.open_dataset(v_file)

u = ds_u['u']
v = ds_v['v']

# ---- SELECT PRESSURE LEVEL (if exists) ----
if 'pressure_level' in u.dims:
    u = u.sel(pressure_level=850)
    v = v.sel(pressure_level=850)

# =========================
# ---- DATE RANGE (EDIT HERE) ----
# =========================
start_date = "2025-10-25"
end_date   = "2025-10-31"

u_sel = u.sel(valid_time=slice(start_date, end_date))
v_sel = v.sel(valid_time=slice(start_date, end_date))

times = u_sel['valid_time'].values
n = len(times)

# =========================
# ---- SUBPLOT GRID ----
# =========================
ncols = min(3, n)
nrows = int(np.ceil(n / ncols))

fig, axes = plt.subplots(
    nrows, ncols,
    figsize=(5*ncols, 4*nrows),
    subplot_kw={'projection': ccrs.PlateCarree()}
)

axes = np.array(axes).flatten()

# =========================
# ---- LOOP ----
# =========================
q_ref = None  # to store first quiver

for i, t in enumerate(times):

    ax = axes[i]

    u_t = u_sel.sel(valid_time=t)
    v_t = v_sel.sel(valid_time=t)

    # ---- INDIA REGION ----
    u_t = u_t.sel(latitude=slice(40, 5), longitude=slice(65, 100))
    v_t = v_t.sel(latitude=slice(40, 5), longitude=slice(65, 100))

    # ---- THIN GRID ----
    skip = 5
    u_plot = u_t[::skip, ::skip]
    v_plot = v_t[::skip, ::skip]

    lon = u_plot.longitude
    lat = u_plot.latitude

    # ---- QUIVER ----
    q = ax.quiver(
        lon, lat,
        u_plot, v_plot,
        scale=200,
        width=0.002,
        headwidth=3
    )

    # store first quiver for legend
    if q_ref is None:
        q_ref = q

    # ---- MAP FEATURES ----
    ax.coastlines()
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    ax.set_extent([65, 100, 5, 40])

    ax.set_title(str(np.datetime_as_string(t, unit='D')))

# =========================
# ---- REMOVE EXTRA AXES ----
# =========================
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

# =========================
# 🔥 REFERENCE ARROW (LEGEND)
# =========================
axes[0].quiverkey(
    q_ref,
    X=0.85, Y=0.95,   # inside top-right
    U=5,
    label="5 m/s",
    labelpos='E'
)
plt.subplots_adjust(bottom=0.15)
# =========================
# ---- FINAL ----
# =========================
plt.tight_layout()
plt.show()
```

```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# ---- LOAD FILE ----
file_path = r"D:\PHD 2025~\1.SERVER_1\Moisture_no_tp_files\climatologies\clim_msl.nc"
ds = xr.open_dataset(file_path)

# ---- SELECT VARIABLE ----
msl = ds['msl']

# ---- SELECT ONE DAY (example: 1st Aug) ----
data = msl.sel(valid_time="2025-08-01")

# ---- SUBSET INDIA REGION ----
# IMPORTANT: latitude is descending → use slice(40, 5)
data_india = data.sel(
    latitude=slice(40, 5),
    longitude=slice(65, 100)
)

# ---- CONVERT Pa → hPa (better for plotting) ----
data_india = data_india / 100.0

# ---- PLOT ----
plt.figure(figsize=(10, 6))

ax = plt.axes(projection=ccrs.PlateCarree())

# contourf plot
cf = ax.contourf(
    data_india.longitude,
    data_india.latitude,
    data_india,
    levels=20,
    cmap="coolwarm",
    transform=ccrs.PlateCarree()
)

# add features
ax.coastlines()
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, facecolor='lightgray', alpha=0.3)

# extent (India region)
ax.set_extent([65, 100, 5, 40], crs=ccrs.PlateCarree())

# colorbar
cbar = plt.colorbar(cf, ax=ax, orientation='vertical', pad=0.02)
cbar.set_label("Mean Sea Level Pressure (hPa)")

# title
plt.title("Climatological MSL (01 Aug)")

plt.show()
```
