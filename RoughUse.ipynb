```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader
from shapely.geometry import box

# ===============================
# 1. DEFINE COORDINATES
# ===============================

# ---- Plot extent (larger box) ----
plot_lon_min, plot_lon_max = 10, 50
plot_lat_min, plot_lat_max = 50, 90

# ---- Study area (smaller box) ----
study_lon_min, study_lon_max = 72, 82
study_lat_min, study_lat_max = 26, 38

# ===============================
# 2. LOAD INDIA SHAPEFILE
# ===============================
# ðŸ‘‰ Change this path to where your India shapefile is stored
india_shp_path = r"/media/nilay/HDD/IIT KGP 2025/PHD/Shapefiles/India Shapefile With Kashmir/India Shape/india_st.shp"

reader = shpreader.Reader(india_shp_path)
india_geom = list(reader.geometries())

# ===============================
# 3. CREATE FIGURE
# ===============================
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection=ccrs.PlateCarree())

# Set plot extent
ax.set_extent(
    [plot_lon_min, plot_lon_max, plot_lat_min, plot_lat_max],
    crs=ccrs.PlateCarree()
)

# ===============================
# 4. ADD INDIA BOUNDARY
# ===============================
ax.add_geometries(
    india_geom,
    crs=ccrs.PlateCarree(),
    facecolor="none",
    edgecolor="black",
    linewidth=1
)

# ===============================
# 5. CREATE BOXES
# ===============================
plot_box = box(plot_lon_min, plot_lat_min, plot_lon_max, plot_lat_max)
study_box = box(study_lon_min, study_lat_min, study_lon_max, study_lat_max)

# Plot extent box (RED)
ax.add_geometries(
    [plot_box],
    crs=ccrs.PlateCarree(),
    facecolor="none",
    edgecolor="red",
    linewidth=2,
    label="Plot extent"
)

# Study area box (BLUE)
ax.add_geometries(
    [study_box],
    crs=ccrs.PlateCarree(),
    facecolor="none",
    edgecolor="blue",
    linewidth=2,
    label="Study area"
)

# ===============================
# 6. GRIDLINES & LABELS
# ===============================
gl = ax.gridlines(
    draw_labels=True,
    linewidth=0.5,
    color="gray",
    alpha=0.5,
    linestyle="--"
)
gl.top_labels = False
gl.right_labels = False

# ===============================
# 7. TITLE
# ===============================
ax.set_title("Study Area within Plot Domain (Latâ€“Lon)", fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader
from shapely.geometry import box

# ===============================
# 1. COORDINATES
# ===============================

# ---- Plot extent (larger box) ----
plot_lon_min, plot_lon_max = 50, 90
plot_lat_min, plot_lat_max = 10, 50

# ---- Study area (smaller box) ----
study_lon_min, study_lon_max = 75, 77
study_lat_min, study_lat_max = 30, 34

# ===============================
# 2. INDIA SHAPEFILE
# ===============================
india_shp_path = r"/media/nilay/HDD/IIT KGP 2025/PHD/Shapefiles/India Shapefile With Kashmir/India Shape/india_st.shp"

reader = shpreader.Reader(india_shp_path)
india_geom = list(reader.geometries())

# ===============================
# 3. FIGURE
# ===============================
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection=ccrs.PlateCarree())

ax.set_extent(
    [plot_lon_min, plot_lon_max, plot_lat_min, plot_lat_max],
    crs=ccrs.PlateCarree()
)

# ===============================
# 4. INDIA BOUNDARY
# ===============================
ax.add_geometries(
    india_geom,
    crs=ccrs.PlateCarree(),
    facecolor="lightgray",
    edgecolor="black",
    linewidth=0.8,
    zorder=2
)

# ===============================
# 5. COASTLINES (CARTOPY)
# ===============================
ax.coastlines(
    resolution="10m",   # "110m", "50m", "10m"
    linewidth=1.0,
    zorder=3
)

# ===============================
# 6. BOXES
# ===============================
plot_box = box(plot_lon_min, plot_lat_min, plot_lon_max, plot_lat_max)
study_box = box(study_lon_min, study_lat_min, study_lon_max, study_lat_max)

# Plot domain (RED)
ax.add_geometries(
    [plot_box],
    crs=ccrs.PlateCarree(),
    facecolor="none",
    edgecolor="red",
    linewidth=2,
    zorder=4
)

# Study area (BLUE)
ax.add_geometries(
    [study_box],
    crs=ccrs.PlateCarree(),
    facecolor="none",
    edgecolor="blue",
    linewidth=2,
    zorder=4
)

# ===============================
# 7. GRIDLINES
# ===============================
gl = ax.gridlines(
    draw_labels=True,
    linestyle="--",
    linewidth=0.5,
    alpha=0.6
)
gl.top_labels = False
gl.right_labels = False

# ===============================
# 8. TITLE
# ===============================
ax.set_title("Study Area within Plot Domain (Latâ€“Lon)", fontsize=12)

plt.show()
```
