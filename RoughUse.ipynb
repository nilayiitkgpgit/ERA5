```python
import cdsapi
import os
import xarray as xr
import glob

# =====================================================
# USER SETTINGS
# =====================================================

out_dir = "/home/nilay/SSD Data Downloads/ERA5_CLIM"
os.makedirs(out_dir, exist_ok=True)

final_file = os.path.join(out_dir, "ERA5_Climatology_1980_2025.nc")

years = range(1998, 1999)
months = ["10"]

# North, West, South, East
area = [50, 60, 0, 100]
#area = None

c = cdsapi.Client()

# =====================================================
# DOWNLOAD SINGLE LEVEL DATA
# =====================================================

single_files = []

for y in years:

    year_str = str(y)
    out_file = os.path.join(out_dir, f"ERA5_single_{year_str}.nc")

    single_files.append(out_file)

    if os.path.exists(out_file):
        print("✅ Skip single:", year_str)
        continue

    print("⬇ Downloading single level:", year_str)

    request = {
        "product_type": "reanalysis",
        "variable": [
            "total_precipitation",
            "vertical_integral_of_divergence_of_moisture_flux",
            "vertical_integral_of_eastward_water_vapour_flux",
            "vertical_integral_of_northward_water_vapour_flux"
        ],
        "year": year_str,
        "month": months,
        "day": [f"{d:02d}" for d in range(1, 32)],
        "daily_statistic": "daily_mean",
        "time_zone": "utc+00:00",
        "frequency": "1_hourly",
        "data_format": "netcdf",
        "area": area
    }

    c.retrieve(
        "derived-era5-single-levels-daily-statistics",
        request,
        out_file
    )

# =====================================================
# DOWNLOAD PRESSURE LEVEL DATA
# =====================================================

pl_files = []

for y in years:

    year_str = str(y)
    out_file = os.path.join(out_dir, f"ERA5_pl_{year_str}.nc")

    pl_files.append(out_file)

    if os.path.exists(out_file):
        print("✅ Skip pressure:", year_str)
        continue

    print("⬇ Downloading pressure level:", year_str)

    request = {
        "product_type": "reanalysis",
        "variable": [
            "geopotential",
            "u_component_of_wind",
            "v_component_of_wind"
        ],
        "pressure_level": "850",
        "year": year_str,
        "month": months,
        "day": [f"{d:02d}" for d in range(1, 32)],
        "daily_statistic": "daily_mean",
        "time_zone": "utc+00:00",
        "frequency": "1_hourly",
        "data_format": "netcdf",
        "area": area
    }

    c.retrieve(
        "derived-era5-pressure-levels-daily-statistics",
        request,
        out_file
    )

# =====================================================
# MERGE ALL FILES
# =====================================================

print("\n📦 Opening datasets...")

ds_single = xr.open_mfdataset(single_files, combine="by_coords")
ds_pl = xr.open_mfdataset(pl_files, combine="by_coords")

print("🔗 Merging datasets...")

ds = xr.merge([ds_single, ds_pl])

print("💾 Saving final file...")

ds.to_netcdf(final_file)

print("\n✅ DONE — Final merged file saved:")
print(final_file)
```

# Checking if the unzipped files are properly cropped or not=>

```python
import xarray as xr
import numpy as np

nc_file = r"D:\PHD 2025~\PHD\5.ERA5_1980-2025_12Months_DailyPrep_unzipped\ERA5_Land_Rainfall_1991.nc"
ds = xr.open_dataset(nc_file)

# ============================================
# GET LAT-LON RANGE
# ============================================

lat = ds['latitude']
lon = ds['longitude']

print("\n🌍 FULL DATA EXTENT:")
print(f"Latitude range  : {float(lat.min())} to {float(lat.max())}")
print(f"Longitude range : {float(lon.min())} to {float(lon.max())}")

# ============================================
# YOUR STUDY REGION
# ============================================

lat_min, lat_max = 20, 32
lon_min, lon_max = 70, 90

# ============================================
# CHECK IF DATA EXCEEDS STUDY REGION
# ============================================

if (lat.min() < lat_min) or (lat.max() > lat_max) or \
   (lon.min() < lon_min) or (lon.max() > lon_max):
    print("\n⚠️ Dataset is LARGER than study region (not cropped)")
else:
    print("\n✅ Dataset is properly cropped to study region")

# ============================================
# CHECK NaNs OUTSIDE STUDY REGION
# ============================================

precip = ds['tp'].isel(valid_time=0)

# Mask INSIDE region
inside = precip.sel(
    latitude=slice(lat_max, lat_min),
    longitude=slice(lon_min, lon_max)
)

# Mask OUTSIDE region
outside = precip.where(
    ( (ds['latitude'] < lat_min) | (ds['latitude'] > lat_max) ) |
    ( (ds['longitude'] < lon_min) | (ds['longitude'] > lon_max) ),
    drop=True
)

print("\n📊 INSIDE REGION:")
print("Min:", float(inside.min()))
print("Max:", float(inside.max()))
print("NaNs:", int(np.isnan(inside.values).sum()))

print("\n📊 OUTSIDE REGION:")
print("Min:", float(outside.min()))
print("Max:", float(outside.max()))
print("NaNs:", int(np.isnan(outside.values).sum()))

# ============================================
# CHECK % OF NaNs OUTSIDE
# ============================================

total_outside = outside.size
nan_outside = np.isnan(outside.values).sum()

print(f"\n📉 Outside NaN Percentage: {100 * nan_outside / total_outside:.2f}%")
```

```python
import xarray as xr
import numpy as np

# Load file
nc_file = r"D:\PHD 2025~\PHD\6.MJO_FILTER\06_MJO_LANCZOS\mjo_anom_ERA5_Land_Rainfall_1984.nc"
ds = xr.open_dataset(nc_file)

print("\n📂 DATASET SUMMARY:")
print(ds)

# ============================================
# CHECK VARIABLE
# ============================================

print("\n📊 VARIABLES:")
print(ds.data_vars)

# Select precipitation
precip = ds['tp']

# ============================================
# CHECK DIMENSIONS
# ============================================

print("\n📐 DIMENSIONS:")
print(precip.dims)

# ============================================
# DETECT TIME DIMENSION (AUTO FIX)
# ============================================

if 'valid_time' in ds.dims:
    time_dim = 'valid_time'
elif 'time' in ds.dims:
    time_dim = 'time'
elif 'dayofyear' in ds.dims:
    time_dim = 'dayofyear'
else:
    raise ValueError("❌ No recognized time dimension found!")

print(f"\n⏱️ Detected time dimension: {time_dim}")

# ============================================
# CHECK TIME VALUES
# ============================================

print("\n⏱️ TIME VALUES:")
print(ds[time_dim].values[:5])

# ============================================
# SELECT ONE TIMESTEP
# ============================================

# Use safe index (avoid out-of-bounds)
index = min(200, ds.dims[time_dim] - 1)

precip_1 = precip.isel({time_dim: index})

print(f"\n📅 Selected index: {index}")

# ============================================
# BASIC STATS
# ============================================

print("\n📈 STATISTICS (RAW):")
print("Min:", float(precip_1.min()))
print("Max:", float(precip_1.max()))
print("Mean:", float(precip_1.mean()))

# Convert to mm
precip_1_mm = precip_1 * 1000

print("\n📈 STATISTICS (mm):")
print("Min:", float(precip_1_mm.min()))
print("Max:", float(precip_1_mm.max()))
print("Mean:", float(precip_1_mm.mean()))

# ============================================
# CHECK FOR NaNs
# ============================================

print("\n❓ NaN CHECK:")
print("Total NaNs:", int(np.isnan(precip_1.values).sum()))

# ============================================
# CHECK REGION VALUES
# ============================================

# Handle latitude direction automatically
lat = ds['latitude']
lon = ds['longitude']

if lat[0] > lat[-1]:  # descending (ERA5 style)
    precip_region = precip_1.sel(
        latitude=slice(32, 20),
        longitude=slice(70, 90)
    )
else:  # ascending
    precip_region = precip_1.sel(
        latitude=slice(20, 32),
        longitude=slice(70, 90)
    )

print("\n📍 REGION STATS:")
print("Min:", float(precip_region.min()))
print("Max:", float(precip_region.max()))
print("Mean:", float(precip_region.mean()))

# ============================================
# PRINT SAMPLE GRID
# ============================================

print("\n🔍 SAMPLE VALUES:")
print(precip_region.values[0:5, 0:5])
```

# Checking the filtered MJO files by plotting=>

```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.io.shapereader as shpreader

# ============================================
# USER INPUT
# ============================================

lat_min, lat_max = 10, 40
lon_min, lon_max = 50, 100

nc_file = r"D:\PHD 2025~\PHD\6.MJO_FILTER\06_MJO_LANCZOS\mjo_anom_ERA5_Land_Rainfall_2025.nc"

shapefile_path = r"D:\PHD 2025~\PHD\Shapefiles\India Shapefile With Kashmir\India Shape\india_st.shp"

extent = [40, 120, 0, 50]

# ============================================
# LOAD DATA
# ============================================

print("📂 Loading dataset...")
ds = xr.open_dataset(nc_file)

print(ds)

# ============================================
# SELECT VARIABLE
# ============================================

precip = ds['tp_mjo']

# Ensure correct dimension order
precip = precip.transpose("time", "latitude", "longitude")

# ============================================
# SELECT ONE DAY (CHANGE INDEX AS NEEDED)
# ============================================

time_index = 200   # 👈 choose your day
precip = precip.isel(time=time_index)

# ============================================
# SUBSET REGION
# ============================================

precip = precip.sel(
    latitude=slice(lat_max, lat_min),   # ERA5 lat is descending
    longitude=slice(lon_min, lon_max)
)

# Fix latitude orientation
precip = precip.sortby('latitude')

# ============================================
# UNIT CONVERSION
# ============================================

precip = precip * 1000   # m → mm

# Optional threshold
precip = precip.where(precip > 1)

# ============================================
# CREATE FIGURE
# ============================================

fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection=ccrs.PlateCarree())

ax.set_extent(extent, crs=ccrs.PlateCarree())

# ============================================
# PLOT
# ============================================

pcm = ax.pcolormesh(
    precip['longitude'],
    precip['latitude'],
    precip,
    cmap='Blues',
    vmin=0,
    vmax=20,
    transform=ccrs.PlateCarree()
)

# Colorbar
cbar = plt.colorbar(pcm, ax=ax, orientation='vertical', pad=0.02)
cbar.set_label("Rainfall (mm/day)")

# ============================================
# MAP FEATURES
# ============================================

ax.coastlines(resolution='110m', linewidth=1)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='white')

# ============================================
# SHAPEFILE
# ============================================

shape = shpreader.Reader(shapefile_path)
ax.add_geometries(
    shape.geometries(),
    crs=ccrs.PlateCarree(),
    facecolor='none',
    edgecolor='black',
    linewidth=1.5
)

# ============================================
# STUDY REGION BOX
# ============================================

lons = [lon_min, lon_max, lon_max, lon_min, lon_min]
lats = [lat_min, lat_min, lat_max, lat_max, lat_min]

ax.plot(
    lons, lats,
    color='red',
    linewidth=2,
    transform=ccrs.PlateCarree(),
    label='Study Region'
)

# ============================================
# GRIDLINES
# ============================================

gl = ax.gridlines(draw_labels=True, linewidth=0.5, linestyle='--')
gl.top_labels = False
gl.right_labels = False

# ============================================
# TITLE
# ============================================

plt.title(f"ERA5 Daily Rainfall (mm) | Time Index = {time_index}", fontsize=14)

plt.legend()
plt.show()
```

```python
import xarray as xr
import pandas as pd
import numpy as np

# ============================================
# USER INPUT (GRID POINT)
# ============================================

lat_point = 25    # 👈 choose your latitude
lon_point = 80    # 👈 choose your longitude

nc_file = r"D:\PHD 2025~\PHD\6.MJO_FILTER\06_MJO_LANCZOS\mjo_anom_ERA5_Land_Rainfall_2025.nc"

# ============================================
# LOAD DATA
# ============================================

print("📂 Loading dataset...")
ds = xr.open_dataset(nc_file)

# Ensure correct dimension order
ds = ds.transpose("time", "latitude", "longitude")

# Extract variable
precip = ds["tp_mjo"] * 1000   # convert to mm

# ============================================
# SELECT GRID POINT (nearest)
# ============================================

point_data = precip.sel(
    latitude=lat_point,
    longitude=lon_point,
    method="nearest"
)

print(f"📍 Selected Grid Point: Lat={float(point_data.latitude.values):.2f}, Lon={float(point_data.longitude.values):.2f}")

# ============================================
# CONVERT TO PANDAS
# ============================================

df = point_data.to_dataframe(name="rainfall").reset_index()

# ============================================
# REMOVE ZERO / NaN VALUES
# ============================================

df = df.dropna()
df = df[df["rainfall"] > 0]   # keep only rainy days

# ============================================
# ADD MONTH COLUMN
# ============================================

df["month"] = df["time"].dt.month

# ============================================
# GROUP BY MONTH
# ============================================

grouped = df.groupby("month")

# ============================================
# PRINT RESULTS
# ============================================

print("\n📊 Available Rainy Days at this Grid Point:\n")

for month, group in grouped:
    print(f"\n📅 Month {month}: ({len(group)} days)")
    
    for t in group["time"]:
        print(f"   {t.strftime('%Y-%m-%d')}")
```
