```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader
from shapely.geometry import box

# ===============================
# 1. DEFINE COORDINATES
# ===============================

# ---- Plot extent (larger box) ----
plot_lon_min, plot_lon_max = 10, 50
plot_lat_min, plot_lat_max = 50, 90

# ---- Study area (smaller box) ----
study_lon_min, study_lon_max = 72, 82
study_lat_min, study_lat_max = 26, 38

# ===============================
# 2. LOAD INDIA SHAPEFILE
# ===============================
# 👉 Change this path to where your India shapefile is stored
india_shp_path = r"/media/nilay/HDD/IIT KGP 2025/PHD/Shapefiles/India Shapefile With Kashmir/India Shape/india_st.shp"

reader = shpreader.Reader(india_shp_path)
india_geom = list(reader.geometries())

# ===============================
# 3. CREATE FIGURE
# ===============================
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection=ccrs.PlateCarree())

# Set plot extent
ax.set_extent(
    [plot_lon_min, plot_lon_max, plot_lat_min, plot_lat_max],
    crs=ccrs.PlateCarree()
)

# ===============================
# 4. ADD INDIA BOUNDARY
# ===============================
ax.add_geometries(
    india_geom,
    crs=ccrs.PlateCarree(),
    facecolor="none",
    edgecolor="black",
    linewidth=1
)

# ===============================
# 5. CREATE BOXES
# ===============================
plot_box = box(plot_lon_min, plot_lat_min, plot_lon_max, plot_lat_max)
study_box = box(study_lon_min, study_lat_min, study_lon_max, study_lat_max)

# Plot extent box (RED)
ax.add_geometries(
    [plot_box],
    crs=ccrs.PlateCarree(),
    facecolor="none",
    edgecolor="red",
    linewidth=2,
    label="Plot extent"
)

# Study area box (BLUE)
ax.add_geometries(
    [study_box],
    crs=ccrs.PlateCarree(),
    facecolor="none",
    edgecolor="blue",
    linewidth=2,
    label="Study area"
)

# ===============================
# 6. GRIDLINES & LABELS
# ===============================
gl = ax.gridlines(
    draw_labels=True,
    linewidth=0.5,
    color="gray",
    alpha=0.5,
    linestyle="--"
)
gl.top_labels = False
gl.right_labels = False

# ===============================
# 7. TITLE
# ===============================
ax.set_title("Study Area within Plot Domain (Lat–Lon)", fontsize=12)

plt.show()
```

```python
import cdsapi
import os
import xarray as xr
import glob

# =====================================================
# USER SETTINGS
# =====================================================

out_dir = "/home/nilay/SSD Data Downloads/ERA5_CLIM"
os.makedirs(out_dir, exist_ok=True)

final_file = os.path.join(out_dir, "ERA5_Climatology_1980_2025.nc")

years = range(1998, 1999)
months = ["10"]

# North, West, South, East
area = [50, 60, 0, 100]
#area = None

c = cdsapi.Client()

# =====================================================
# DOWNLOAD SINGLE LEVEL DATA
# =====================================================

single_files = []

for y in years:

    year_str = str(y)
    out_file = os.path.join(out_dir, f"ERA5_single_{year_str}.nc")

    single_files.append(out_file)

    if os.path.exists(out_file):
        print("✅ Skip single:", year_str)
        continue

    print("⬇ Downloading single level:", year_str)

    request = {
        "product_type": "reanalysis",
        "variable": [
            "total_precipitation",
            "vertical_integral_of_divergence_of_moisture_flux",
            "vertical_integral_of_eastward_water_vapour_flux",
            "vertical_integral_of_northward_water_vapour_flux"
        ],
        "year": year_str,
        "month": months,
        "day": [f"{d:02d}" for d in range(1, 32)],
        "daily_statistic": "daily_mean",
        "time_zone": "utc+00:00",
        "frequency": "1_hourly",
        "data_format": "netcdf",
        "area": area
    }

    c.retrieve(
        "derived-era5-single-levels-daily-statistics",
        request,
        out_file
    )

# =====================================================
# DOWNLOAD PRESSURE LEVEL DATA
# =====================================================

pl_files = []

for y in years:

    year_str = str(y)
    out_file = os.path.join(out_dir, f"ERA5_pl_{year_str}.nc")

    pl_files.append(out_file)

    if os.path.exists(out_file):
        print("✅ Skip pressure:", year_str)
        continue

    print("⬇ Downloading pressure level:", year_str)

    request = {
        "product_type": "reanalysis",
        "variable": [
            "geopotential",
            "u_component_of_wind",
            "v_component_of_wind"
        ],
        "pressure_level": "850",
        "year": year_str,
        "month": months,
        "day": [f"{d:02d}" for d in range(1, 32)],
        "daily_statistic": "daily_mean",
        "time_zone": "utc+00:00",
        "frequency": "1_hourly",
        "data_format": "netcdf",
        "area": area
    }

    c.retrieve(
        "derived-era5-pressure-levels-daily-statistics",
        request,
        out_file
    )

# =====================================================
# MERGE ALL FILES
# =====================================================

print("\n📦 Opening datasets...")

ds_single = xr.open_mfdataset(single_files, combine="by_coords")
ds_pl = xr.open_mfdataset(pl_files, combine="by_coords")

print("🔗 Merging datasets...")

ds = xr.merge([ds_single, ds_pl])

print("💾 Saving final file...")

ds.to_netcdf(final_file)

print("\n✅ DONE — Final merged file saved:")
print(final_file)
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.io.shapereader as shpreader

# ============================================
# USER INPUT
# ============================================

# Study region (EDIT THIS)
lat_min, lat_max = 20, 32
lon_min, lon_max = 70, 90

# Shapefile path (India boundary or basin)
shapefile_path = r"D:\PHD 2025~\PHD\Shapefiles\India Shapefile With Kashmir\India Shape\india_st.shp"

# Map extent (bigger than box)
extent = [40, 120, 0, 50]   # [lon_min, lon_max, lat_min, lat_max]

# ============================================
# CREATE FIGURE
# ============================================

fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection=ccrs.PlateCarree())

# ============================================
# SET MAP EXTENT
# ============================================

ax.set_extent(extent, crs=ccrs.PlateCarree())

# ============================================
# ADD FEATURES
# ============================================

ax.coastlines(resolution='110m', linewidth=1)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='white')

# ============================================
# ADD SHAPEFILE
# ============================================

shape = shpreader.Reader(shapefile_path)
ax.add_geometries(
    shape.geometries(),
    crs=ccrs.PlateCarree(),
    facecolor='none',
    edgecolor='black',
    linewidth=1.5
)

# ============================================
# DRAW LAT-LON BOX
# ============================================

# Box corners
lons = [lon_min, lon_max, lon_max, lon_min, lon_min]
lats = [lat_min, lat_min, lat_max, lat_max, lat_min]

ax.plot(
    lons, lats,
    color='red',
    linewidth=2,
    transform=ccrs.PlateCarree(),
    label='Study Region'
)

# ============================================
# GRIDLINES
# ============================================

gl = ax.gridlines(draw_labels=True, linewidth=0.5, linestyle='--')
gl.top_labels = False
gl.right_labels = False

# ============================================
# TITLE
# ============================================

plt.title("Study Region over India", fontsize=14)

plt.legend()
plt.show()
```
